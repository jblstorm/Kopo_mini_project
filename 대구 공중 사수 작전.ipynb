```python
import pygame
import random
import sys

# 초기화
pygame.init()

# 화면 크기 설정
screen_width = 800
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))

# 게임 제목 설정
pygame.display.set_caption("대구 사수 작전")

# FPS 설정
clock = pygame.time.Clock()

# 이미지 로드
background = pygame.image.load(r'C:\Users\ORG\Desktop\배경.jpg').convert()
fighter_img = pygame.image.load(r'C:\Users\ORG\Desktop\전투기예제.png')
enemy_img = pygame.image.load(r'C:\Users\ORG\Desktop\mig-29.png')  # 적 기체 이미지
bomb_img = pygame.image.load(r'C:\Users\ORG\Desktop\항공폭탄.png')  # 폭탄 이미지

# 전투기 초기 위치
fighter_x = screen_width / 2
fighter_y = screen_height - 70

# 적 기체 리스트
enemies = []
bombs = []  # 폭탄 리스트

# 게임 점수
score = 0
font = pygame.font.SysFont("arial", 36)

# 게임 제한 시간 (초)
game_time = 60

# 적 기체 추가
def add_enemy():
    x = random.randrange(0, screen_width - enemy_img.get_width())
    y = -enemy_img.get_height()
    direction = random.choice(['left', 'right', 'up', 'down'])
    enemies.append([x, y, direction])

# 폭탄 발사
def fire_bomb(x, y, direction):
    bombs.append([x, y, direction])

# 적 기체 움직임 처리
def handle_enemies(enemies):
    for enemy in enemies[:]:
        # 움직임 방향 변경
        if random.randint(1, 60) == 1:  # 1/60 확률로 방향 변경
            enemy[2] = random.choice(['left', 'right', 'up', 'down'])

        # 현재 방향으로 움직임 처리
        if enemy[2] == 'left' and enemy[0] > 0:
            enemy[0] -= 3
        elif enemy[2] == 'right' and enemy[0] < screen_width - enemy_img.get_width():
            enemy[0] += 3
        elif enemy[2] == 'up' and enemy[1] > 0:
            enemy[1] -= 3
        elif enemy[2] == 'down' and enemy[1] < screen_height - enemy_img.get_height():
            enemy[1] += 3

        # 화면에 적 그리기
        screen.blit(enemy_img, (enemy[0], enemy[1]))

# 폭탄 움직임 처리
def handle_bombs(bombs):
    for bomb in bombs[:]:
        if bomb[2] == 'up':
            bomb[1] -= 10
        elif bomb[2] == 'down':
            bomb[1] += 10
        elif bomb[2] == 'left':
            bomb[0] -= 10
        elif bomb[2] == 'right':
            bomb[0] += 10
        
        # 폭탄이 화면 바깥으로 나가면 제거
        if (bomb[1] < -bomb_img.get_height() or bomb[1] > screen_height or
            bomb[0] < -bomb_img.get_width() or bomb[0] > screen_width):
            bombs.remove(bomb)
        else:
            screen.blit(bomb_img, (bomb[0], bomb[1]))

# 충돌 감지
def check_collision(fighter_x, fighter_y, enemies, bombs):
    global score
    fighter_rect = pygame.Rect(fighter_img.get_rect(topleft=(fighter_x, fighter_y)))
    for enemy in enemies[:]:
        enemy_rect = pygame.Rect(enemy_img.get_rect(topleft=(enemy[0], enemy[1])))
        for bomb in bombs[:]:
            bomb_rect = pygame.Rect(bomb_img.get_rect(topleft=(bomb[0], bomb[1])))
            if bomb_rect.colliderect(enemy_rect):
                enemies.remove(enemy)
                bombs.remove(bomb)
                score += 100  # 점수 증가
                break

# 게임 루프
running = True
enemy_timer = 0
bomb_timer = 0
start_ticks = pygame.time.get_ticks()  # 타이머 시작
last_direction = 'up'  # 기본 방향

while running:
    elapsed_time = (pygame.time.get_ticks() - start_ticks) / 1000  # 경과 시간 계산
    if elapsed_time > game_time:
        print(f"Game Over! Your Score: {score}")
        pygame.quit()
        sys.exit()

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                if bomb_timer >= 15:  # 폭탄 발사 간격 0.5초
                    fire_bomb(fighter_x + fighter_img.get_width() // 2 - bomb_img.get_width() // 2, fighter_y, last_direction)
                    bomb_timer = 0

    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and fighter_x > 0:
        fighter_x -= 5
        last_direction = 'left'
    if keys[pygame.K_RIGHT] and fighter_x < screen_width - fighter_img.get_width():
        fighter_x += 5
        last_direction = 'right'
    if keys[pygame.K_UP] and fighter_y > 0:
        fighter_y -= 5
        last_direction = 'up'
    if keys[pygame.K_DOWN] and fighter_y < screen_height - fighter_img.get_height():
        fighter_y += 5
        last_direction = 'down'

    screen.blit(background, (0, 0))
    handle_enemies(enemies)
    handle_bombs(bombs)
    check_collision(fighter_x, fighter_y, enemies, bombs)
    score_text = font.render(f"Score: {score}", True, (255, 255, 255))
    screen.blit(score_text, (10, 10))
    screen.blit(fighter_img, (fighter_x, fighter_y))
    pygame.display.flip()

    # 적 기체 생성 타이머 (10초마다)
    if enemy_timer >= 300:  # 10초마다 적 기체 추가 (FPS 30)
        add_enemy()
        enemy_timer = 0

    enemy_timer += 1
    bomb_timer += 1
    clock.tick(30)

pygame.quit()
sys.exit()
```
